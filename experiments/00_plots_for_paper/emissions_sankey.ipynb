```python
import os

import plotly.graph_objects as go
import pandas as pd
from energyemissionsregio.config import DATA_PATH
```

```python
iron_and_steel_industries = pd.read_csv(os.path.join(DATA_PATH, "ghg_emissions_from_fc_in_iron_and_steel_industries.csv"))
non_ferrous_metals_industries = pd.read_csv(os.path.join(DATA_PATH, "ghg_emissions_from_fc_in_non_ferrous_metals_industries.csv"))
chemical_industries = pd.read_csv(os.path.join(DATA_PATH, "ghg_emissions_from_fc_in_chemical_industries.csv"))
non_metallic_minerals_industries = pd.read_csv(os.path.join(DATA_PATH, "ghg_emissions_from_fc_in_non_metallic_minerals_industries.csv"))
pulp_paper_and_printing_industries = pd.read_csv(os.path.join(DATA_PATH, "ghg_emissions_from_fc_in_pulp_paper_and_printing_industries.csv"))
food_beverages_and_tobacco_industries = pd.read_csv(os.path.join(DATA_PATH, "ghg_emissions_from_fc_in_food_beverages_and_tobacco_industries.csv"))
other_manufacturing_industries_and_construction = pd.read_csv(os.path.join(DATA_PATH, "ghg_emissions_from_fc_in_other_manufacturing_industries_and_construction.csv"))
rail_transport = pd.read_csv(os.path.join(DATA_PATH, "ghg_emissions_from_fc_in_rail_transport.csv"))

light_duty_trucks = pd.read_csv(os.path.join(DATA_PATH, "ghg_emissions_from_fc_in_road_transport_using_light_duty_trucks.csv"))
heavy_duty_trucks_and_buses = pd.read_csv(os.path.join(DATA_PATH, "ghg_emissions_from_fc_in_road_transport_using_heavy_duty_trucks_and_buses.csv"))
freight = pd.merge(light_duty_trucks, heavy_duty_trucks_and_buses, on="region_code")
freight["value"] = freight["value_x"] + freight["value_y"]

cars = pd.read_csv(os.path.join(DATA_PATH, "ghg_emissions_from_fc_in_road_transport_using_cars.csv"))
motorcycles = pd.read_csv(os.path.join(DATA_PATH, "ghg_emissions_from_fc_in_road_transport_using_motorcycles.csv"))

domestic_aviation = pd.read_csv(os.path.join(DATA_PATH, "ghg_emissions_from_fc_in_domestic_aviation.csv"))
domestic_navigation = pd.read_csv(os.path.join(DATA_PATH, "ghg_emissions_from_fc_in_domestic_navigation.csv"))

households = pd.read_csv(os.path.join(DATA_PATH, "ghg_emissions_from_fc_in_households.csv"))
commerce = pd.read_csv(os.path.join(DATA_PATH, "ghg_emissions_from_fc_in_commerce.csv"))

sum_vars = ["ghg_emissions_in_agriculture_from_rice_cultivation", 
            "ghg_emissions_in_agriculture_from_managed_agricultural_soils",
            "ghg_emissions_in_agriculture_from_prescribed_burning_of_savannas",
            "ghg_emissions_in_agriculture_from_field_burning_of_agricultural_residues",
            "ghg_emissions_in_agriculture_from_liming",
            "ghg_emissions_in_agriculture_from_urea_application",
            "ghg_emissions_in_agriculture_from_other_carbon_containing_fertilizers"]

cultivation = None
for var in sum_vars:
    temp_df = pd.read_csv(os.path.join(DATA_PATH, f"{var}.csv"))
    temp_df.drop(columns="year", inplace=True)

    if cultivation is None:
        cultivation = temp_df
    else:
        cultivation = pd.merge(cultivation, temp_df, on="region_code")
        cultivation["value"] = cultivation["value_x"] + cultivation["value_y"]
        cultivation.drop(columns=["value_x", "value_y"], inplace=True)

sum_vars = ["ghg_emissions_in_agriculture_from_enteric_fermentation_of_cattle",
            "ghg_emissions_in_agriculture_from_enteric_fermentation_of_sheep",
            "ghg_emissions_in_agriculture_from_enteric_fermentation_of_swine",
            "ghg_emissions_in_agriculture_from_enteric_fermentation_of_other_livestock",
            "ghg_emissions_in_agriculture_from_cattle_manure_management",
            "ghg_emissions_in_agriculture_from_sheep_manure_management",
            "ghg_emissions_in_agriculture_from_swine_manure_management",
            "ghg_emissions_in_agriculture_from_other_livestock_manure_management",
            "ghg_emissions_in_agriculture_from_manure_management_indirect_n20_emissions"]

livestock = None
for var in sum_vars:
    temp_df = pd.read_csv(os.path.join(DATA_PATH, f"{var}.csv"))
    temp_df.drop(columns="year", inplace=True)

    if livestock is None:
        livestock = temp_df
    else:
        livestock = pd.merge(livestock, temp_df, on="region_code")
        livestock["value"] = livestock["value_x"] + livestock["value_y"]
        livestock.drop(columns=["value_x", "value_y"], inplace=True)
```

```python
iron_and_steel_industries_de = iron_and_steel_industries[iron_and_steel_industries["region_code"] == "DE"]["value"].item()
iron_and_steel_industries_es = iron_and_steel_industries[iron_and_steel_industries["region_code"] == "ES"]["value"].item()

non_ferrous_metals_industries_de = non_ferrous_metals_industries[non_ferrous_metals_industries["region_code"] == "DE"]["value"].item()
non_ferrous_metals_industries_es = non_ferrous_metals_industries[non_ferrous_metals_industries["region_code"] == "ES"]["value"].item()

chemical_industries_de = chemical_industries[chemical_industries["region_code"] == "DE"]["value"].item()
chemical_industries_es = chemical_industries[chemical_industries["region_code"] == "ES"]["value"].item()

non_metallic_minerals_industries_de = non_metallic_minerals_industries[non_metallic_minerals_industries["region_code"] == "DE"]["value"].item()
non_metallic_minerals_industries_es = non_metallic_minerals_industries[non_metallic_minerals_industries["region_code"] == "ES"]["value"].item()

pulp_paper_and_printing_industries_de = pulp_paper_and_printing_industries[pulp_paper_and_printing_industries["region_code"] == "DE"]["value"].item()
pulp_paper_and_printing_industries_es = pulp_paper_and_printing_industries[pulp_paper_and_printing_industries["region_code"] == "ES"]["value"].item()

food_beverages_and_tobacco_industries_de = food_beverages_and_tobacco_industries[food_beverages_and_tobacco_industries["region_code"] == "DE"]["value"].item()
food_beverages_and_tobacco_industries_es = food_beverages_and_tobacco_industries[food_beverages_and_tobacco_industries["region_code"] == "ES"]["value"].item()

other_manufacturing_industries_and_construction_de = other_manufacturing_industries_and_construction[other_manufacturing_industries_and_construction["region_code"] == "DE"]["value"].item()
other_manufacturing_industries_and_construction_es = other_manufacturing_industries_and_construction[other_manufacturing_industries_and_construction["region_code"] == "ES"]["value"].item()

rail_transport_de = rail_transport[rail_transport["region_code"] == "DE"]["value"].item()
rail_transport_es = rail_transport[rail_transport["region_code"] == "ES"]["value"].item()

cars_de = cars[cars["region_code"] == "DE"]["value"].item()
cars_es = cars[cars["region_code"] == "ES"]["value"].item()

motorcycles_de = motorcycles[motorcycles["region_code"] == "DE"]["value"].item()
motorcycles_es = motorcycles[motorcycles["region_code"] == "ES"]["value"].item()

freight_de = freight[freight["region_code"] == "DE"]["value"].item()
freight_es = freight[freight["region_code"] == "ES"]["value"].item()

domestic_aviation_de = domestic_aviation[domestic_aviation["region_code"] == "DE"]["value"].item()
domestic_aviation_es = domestic_aviation[domestic_aviation["region_code"] == "ES"]["value"].item()

domestic_navigation_de = domestic_navigation[domestic_navigation["region_code"] == "DE"]["value"].item()
domestic_navigation_es = domestic_navigation[domestic_navigation["region_code"] == "ES"]["value"].item()

households_de = households[households["region_code"] == "DE"]["value"].item()
households_es = households[households["region_code"] == "ES"]["value"].item()

commerce_de = commerce[commerce["region_code"] == "DE"]["value"].item()
commerce_es = commerce[commerce["region_code"] == "ES"]["value"].item()

livestock_de = livestock[livestock["region_code"] == "DE"]["value"].item()
livestock_es = livestock[livestock["region_code"] == "ES"]["value"].item()

cultivation_de = cultivation[cultivation["region_code"] == "DE"]["value"].item()
cultivation_es = cultivation[cultivation["region_code"] == "ES"]["value"].item()
```

```python
de_emissions = [iron_and_steel_industries_de, non_ferrous_metals_industries_de, chemical_industries_de, non_metallic_minerals_industries_de, 
                pulp_paper_and_printing_industries_de, food_beverages_and_tobacco_industries_de, other_manufacturing_industries_and_construction_de,
                rail_transport_de, cars_de, motorcycles_de, freight_de, domestic_aviation_de, domestic_navigation_de, 
                households_de, commerce_de, livestock_de, cultivation_de]

es_emissions = [iron_and_steel_industries_es, non_ferrous_metals_industries_es, chemical_industries_es, non_metallic_minerals_industries_es, 
                pulp_paper_and_printing_industries_es, food_beverages_and_tobacco_industries_es, other_manufacturing_industries_and_construction_es,
                rail_transport_es, cars_es, motorcycles_es, freight_es, domestic_aviation_es, domestic_navigation_es, 
                households_es, commerce_es, livestock_es, cultivation_es]
```

```python
df_emissions_de = pd.DataFrame({
    'sector': sector,
    'sub_sector': sub_sector,
    'end_use_sector': end_use_sector,
    'value': de_emissions
})

df_emissions_es = pd.DataFrame({
    'sector': sector,
    'sub_sector': sub_sector,
    'end_use_sector': end_use_sector,
    'value': es_emissions
})
```

```python
# Define color mapping
sector_colors = {
    "Industry": "#1f77b4",  # Blue
    "Transport": "#ff7f0e",  # Orange
    "Households": "#2ca02c",  # Green
    "Commerce": "#d62728",  # Red
    "Agriculture": "#9467bd",  # Purple

    "Energy-intensive": "#1f77b4",  # Blue
    "Non-energy-intensive": "#1f77b4",  # Blue

    "Rail": "#ff7f0e",  # Orange
    "Road": "#ff7f0e",  # Orange
    "Domestic aviation": "#ff7f0e",  # Orange 
    "Domestic navigation": "#ff7f0e",  # Orange
    "Livestock": "#9467bd",  # Purple 
    "Cultivation": "#9467bd",  # Purple 

    "Iron and steel": "#1f77b4",  # Blue
    "Non-ferrous metals": "#1f77b4",  # Blue
    "Chemical": "#1f77b4",  # Blue
     "Non-metallic minerals": "#1f77b4",  # Blue
    "Paper, pulp, and printing": "#1f77b4",  # Blue
    "Food, beverages, and tobacco": "#1f77b4",  # Blue
    "Other manufacturing industries and construction": "#1f77b4",  # Blue

    "Cars": "#ff7f0e",  # Orange
      "Motorcycles": "#ff7f0e",  # Orange
        "Freight": "#ff7f0e",  # Orange

}
```

```python
df_emissions_de.to_csv(os.path.join("..", "..", "data", "NUTS0_data", "emissions_de.csv"), index=False)
df_emissions_es.to_csv(os.path.join("..", "..", "data", "NUTS0_data", "emissions_es.csv"), index=False)
```

```python
# Get all unique nodes
node_labels = list(pd.unique(df_emissions_de[['sector', 'sub_sector', 'end_use_sector']].values.ravel('K')))
node_indices = {label: i for i, label in enumerate(node_labels)}

# Create source-target relationships for both transitions
source = df_emissions_de['sector'].map(node_indices).tolist() + df_emissions_de['sub_sector'].map(node_indices).tolist()
target = df_emissions_de['sub_sector'].map(node_indices).tolist() + df_emissions_de['end_use_sector'].map(node_indices).tolist()

values_de = df_emissions_de['value'].tolist() + df_emissions_de['value'].tolist()  
values_es = df_emissions_es['value'].tolist() + df_emissions_es['value'].tolist()  

# Create mapping for node indices
node_indices = {label: i for i, label in enumerate(node_labels)}

# Assign colors
node_colors = []
for label in node_labels:
    if label in sector_colors:
        node_colors.append(sector_colors[label])
    else:
        node_colors.append("#7f7f7f")  # Grey for unspecified breakdowns
    
# Germany
sankey_de = go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=node_labels,
        color=node_colors
    ),
    link=dict(
        color='#ffeda0',
        source=source,
        target=target,
        value=values_de
    )
)

# Second Sankey diagram (Duplicate for now, but can be adjusted for different data)
sankey_es = go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=node_labels,
        color=node_colors
    ),
    link=dict(
        color='#ffeda0',
        source=source,
        target=target,
        value=values_es
    )
)

# Create figure with subplots
from plotly.subplots import make_subplots

fig = make_subplots(rows=2, cols=1, 
                    subplot_titles=["<b>Germany</b>", "<b>Spain</b>"],
                    specs=[[{"type": "sankey"}], [{"type": "sankey"}]],
                    vertical_spacing=0.07)  # Reduce white space between sankeys

fig.add_trace(sankey_de, row=1, col=1)
fig.add_trace(sankey_es, row=2, col=1)

fig.update_layout(
    autosize=False,
    width=1700,
    height=1300,
    font=dict(size=18),
)

fig.write_html(os.path.join("..", "..", "figures", "emissions_categorisation.html")) #NOTE: could not save as png. so saved as HTML and then manually saved as png.

fig.show()
```
