```python
df_fec_de.to_csv(os.path.join("..", "..", "data", "NUTS0_data", "fec_de.csv"), index=False)
df_fec_es.to_csv(os.path.join("..", "..", "data", "NUTS0_data", "fec_es.csv"), index=False)
```

```python
# Get all unique nodes
node_labels = list(pd.unique(df_fec_de[['sector', 'sub_sector', 'end_use_sector']].values.ravel('K')))
node_indices = {label: i for i, label in enumerate(node_labels)}

# Create source-target relationships for both transitions
source = df_fec_de['sector'].map(node_indices).tolist() + df_fec_de['sub_sector'].map(node_indices).tolist()
target = df_fec_de['sub_sector'].map(node_indices).tolist() + df_fec_de['end_use_sector'].map(node_indices).tolist()

values_de = df_fec_de['value'].tolist() + df_fec_de['value'].tolist()  
values_es = df_fec_es['value'].tolist() + df_fec_es['value'].tolist()  

# Create mapping for node indices
node_indices = {label: i for i, label in enumerate(node_labels)}

# Assign colors
node_colors = []
for label in node_labels:
    if label in sector_colors:
        node_colors.append(sector_colors[label])
    else:
        node_colors.append("#7f7f7f")  # Grey for unspecified breakdowns
    
# Germany
sankey_de = go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=node_labels,
        color=node_colors
    ),
    link=dict(
        color='#ffeda0',
        source=source,
        target=target,
        value=values_de
    )
)

# Second Sankey diagram (Duplicate for now, but can be adjusted for different data)
sankey_es = go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=node_labels,
        color=node_colors
    ),
    link=dict(
        color='#ffeda0',
        source=source,
        target=target,
        value=values_es
    )
)

# Create figure with subplots
from plotly.subplots import make_subplots

fig = make_subplots(rows=2, cols=1, 
                    subplot_titles=["<b>Germany</b>", "<b>Spain</b>"],
                    specs=[[{"type": "sankey"}], [{"type": "sankey"}]],
                    vertical_spacing=0.07)  # Reduce white space between sankeys

fig.add_trace(sankey_de, row=1, col=1)
fig.add_trace(sankey_es, row=2, col=1)

fig.update_layout(
    autosize=False,
    width=1700,
    height=1300,
    font=dict(size=18),
)

fig.write_html(os.path.join("..", "..", "figures", "fec_categorisation.html")) #NOTE: could not save as png. so saved as HTML and then manually saved as png.

fig.show()
```
